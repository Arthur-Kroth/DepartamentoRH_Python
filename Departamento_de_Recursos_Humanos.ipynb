#Técnicas de aprendizado de máquina

As técnicas de aprendizado de máquina que serão mencionadas abaixo (regressão linear, regressão logística, random forest e redes neurais artificiais) são classificadas como "técnicas de aprendizado de máquina". Cada uma delas é usada para resolver problemas específicos de aprendizado, seja previsão de valores contínuos (regressão) ou classificação em categorias (classificação).

Essas técnicas de aprendizado de máquina permitem que os algoritmos analisem dados, encontrem padrões e tomem decisões com base nesses padrões, permitindo que os modelos façam previsões ou classificações em novos dados não vistos anteriormente.

Portanto, são ferramentas poderosas para desenvolver modelos de previsão em uma ampla variedade de áreas, como análise de dados, reconhecimento de padrões, diagnósticos médicos, previsão de vendas, entre muitos outros.

###Regressão Linear:
A regressão linear é um método estatístico usado para encontrar uma relação linear simples entre uma variável dependente (o que queremos prever) e uma ou mais variáveis independentes (características que ajudam a fazer a previsão). É aplicada quando queremos fazer previsões contínuas, como estimar o preço de uma casa com base em sua área e número de quartos.

Fórmula: **y = mx + b**

Na regressão linear simples, a fórmula matemática é uma equação de uma linha reta chamada de equação linear (y = mx + b), onde "y" representa a variável dependente (o que queremos prever), "x" é a variável independente (a característica usada para a previsão), "m" é o coeficiente angular (representa a inclinação da linha) e "b" é o coeficiente linear (representa o ponto de interseção da linha com o eixo y).

O objetivo é encontrar os valores de "m" e "b" que melhor se ajustam aos dados, minimizando o erro entre os valores previstos pela linha reta e os valores reais observados.

###Regressão logística

A regressão logística é um método usado para fazer previsões em problemas de classificação binária, onde queremos classificar uma amostra em duas categorias (por exemplo, sim/não, positivo/negativo). Ela estima a probabilidade de uma amostra pertencer a uma das classes. É aplicada em cenários como detecção de spam em emails ou previsão de sucesso/falha em um projeto.

Fórmula: **p = 1 / (1 + e^(-z))**

Na regressão logística, o objetivo é estimar a probabilidade de uma amostra pertencer a uma das duas classes (classificação binária).

A fórmula matemática envolve o uso de uma função logística, também conhecida como função sigmoide (p = 1 / (1 + e^(-z))), onde "p" é a probabilidade estimada e "z" é o resultado da soma ponderada das características da amostra multiplicadas pelos coeficientes do modelo.

Os coeficientes são encontrados durante o treinamento do modelo, ajustando-os para maximizar a probabilidade correta das classes conhecidas.

##Matriz de Confusão

Uma matriz de confusão é uma tabela que é comumente usada para avaliar o desempenho de um modelo de classificação. Ela mostra a relação entre as previsões feitas pelo modelo e as classes reais dos dados.

A matriz de confusão é composta por quatro valores principais:

1. Verdadeiro Positivo (TP): Representa o número de instâncias corretamente classificadas como positivas (classe verdadeira) pelo modelo.

2. Verdadeiro Negativo (TN): Representa o número de instâncias corretamente classificadas como negativas (classe falsa) pelo modelo.

3. Falso Positivo (FP): Representa o número de instâncias erroneamente classificadas como positivas (classe verdadeira) pelo modelo.

4. Falso Negativo (FN): Representa o número de instâncias erroneamente classificadas como negativas (classe falsa) pelo modelo.

Essa informação é organizada em uma tabela, onde as classes reais são representadas nas linhas e as classes previstas pelo modelo são representadas nas colunas. A matriz de confusão geralmente tem o seguinte formato:

|              | Classe Positiva (verdadeira) | Classe Negativa (falsa) |
|--------------|----------------------------|-------------------------|
| Positivo     | Verdadeiro Positivo (TP)    | Falso Positivo (FP)     |
| Negativo     | Falso Negativo (FN)         | Verdadeiro Negativo (TN)|

Como funciona:

Para calcular a matriz de confusão, o modelo é testado em um conjunto de dados de teste, cujas classes verdadeiras são conhecidas. Cada amostra do conjunto de teste é classificada pelo modelo e, em seguida, comparada com sua classe real. Com base nas classificações corretas e incorretas, os valores TP, TN, FP e FN são contabilizados na matriz de confusão.

Exemplo:

Vamos considerar um exemplo de um modelo de detecção de spam em e-mails. Suponha que temos um conjunto de 100 e-mails, onde 50 são spam e 50 são e-mails legítimos (não spam).

O modelo faz suas previsões e os resultados são os seguintes:

|              | Classe Spam | Classe Não Spam |
|--------------|-------------|-----------------|
| Spam         | 45 (TP)     | 5 (FP)          |
| Não Spam     | 8 (FN)      | 42 (TN)         |

Neste exemplo, o modelo acertou 45 e-mails spam (TP) e 42 e-mails não spam (TN), mas errou 8 e-mails spam, classificando-os como não spam (FN), e errou 5 e-mails não spam, classificando-os como spam (FP).

A matriz de confusão nos ajuda a ter uma visão clara do desempenho do modelo, mostrando os acertos e os erros em suas previsões, e pode ser usada para calcular várias métricas de avaliação, como a acurácia, precisão, recall, entre outras.

#Métricas de avaliação *Precision* e *Recall*
Precision e Recall, também conhecidos como Sensibilidade, são métricas de avaliação amplamente utilizadas em problemas de classificação, incluindo a regressão logística. Essas métricas fornecem informações sobre a qualidade das previsões feitas pelo modelo em relação a uma classe específica.

##Precision (Precisão):
A precisão é a proporção de instâncias corretamente classificadas como positivas (verdadeiros positivos) em relação a todas as instâncias que o modelo classificou como positivas (verdadeiros positivos + falsos positivos).

Em outras palavras, é a capacidade do modelo de não classificar erroneamente uma amostra como positiva quando ela é na verdade negativa.


**Fórmula de Cálculo da Precision:**

Precision = TP / (TP + FP)

*Como nossa intenção é prever os funcionários que irão sair da empresa, tivemos que olhar para o nosso modelo treinado de outra maneira, porém seguindo a lógica que deverá ser aplicada a qualquer outro modelo:*

TP (True Positive): O número de instâncias corretamente classificadas como positivas.

FP (False Positive): O número de instâncias erroneamente classificadas como positivas (quando na verdade são negativas).

FN (False Negative): O número de instâncias erroneamente classificadas como negativas (quando na verdade são positivas).
___

##Recall:
O recall é a proporção de instâncias corretamente classificadas como positivas (verdadeiros positivos) em relação a todas as instâncias que são realmente positivas (verdadeiros positivos + falsos negativos). Em outras palavras, é a capacidade do modelo de identificar todas as instâncias positivas corretamente.

Fórmula de Cálculo do Recall:

Recall = TP / (TP + FN)

## Random forest
Random Forest (floresta aleatória) é um algoritmo de aprendizado de máquina que combina várias árvores de decisão para obter um resultado final mais robusto e preciso. É aplicado em problemas de classificação e regressão. É útil quando se deseja evitar overfitting (ajuste excessivo aos dados de treinamento) e para lidar com dados de alta dimensionalidade. É amplamente utilizado em áreas como detecção de fraudes, diagnósticos médicos e previsão de vendas.

A fórmula matemática de cada árvore de decisão envolve uma série de testes (decisões) em cada nó da árvore, comparando as características da amostra com valores de corte (thresholds) para decidir qual caminho seguir até alcançar as folhas (os resultados). As previsões de todas as árvores são combinadas usando votação (no caso de classificação) ou média (no caso de regressão) para chegar ao resultado final.

## Redes neurais artificiais

- Quantos neurônios e quantas camadas? https://iaexpert.academy/2020/05/04/quantas-camadas-escondidas-e-quantos-neuronios-incluir-numa-rede-neural-artificial/

Redes neurais são modelos inspirados no funcionamento do cérebro humano. Elas consistem em camadas de neurônios artificiais que processam informações para fazer previsões ou tomar decisões. As redes neurais são aplicadas em uma ampla gama de tarefas de aprendizado de máquina, como reconhecimento de imagens, processamento de linguagem natural, previsão do mercado financeiro e jogos.

As redes neurais são compostas por camadas de neurônios conectados. A fórmula matemática para um neurônio geralmente envolve uma combinação linear das entradas (características da amostra) ponderadas pelos pesos sinápticos, seguida da aplicação de uma função de ativação.

A fórmula é geralmente representada como: saída = função_ativação(soma_ponderada_entradas).

Existem diferentes funções de ativação, como a função sigmoide, a função ReLU (unidade retificadora linear) e outras, que introduzem não linearidades na rede, permitindo a modelagem de relações complexas nos dados.

##Funções de ativação

RELU:
Esta é a principal função de ativação especialmente em cenários mais complexos como em deep leraning.
Se o valor for maior do que zero a função retorna o próprio valor, caso contrário ela retornará zero.

SIGMOID:
A função Sigmoid é limitada no intervalo (0, 1), o que é útil para interpretar suas saídas como probabilidades.
Se o valor de saída da função Sigmoid for maior ou igual a 0,5, a amostra é classificada como verdadeira (ou pertencente à classe positiva). Se o valor for menor que 0,5, a amostra é classificada como falsa (ou pertencente à classe negativa).

##Compilando nossa rede neural
Nesta fase, são definidas algumas configurações importantes que determinam como o modelo será treinado e como a performance será avaliada durante o treinamento.

###Aqui está o significado de cada argumento usado na função compile:


*   **optimizer='Adam':**
O otimizador é o algoritmo responsável por ajustar os pesos da rede neural durante o treinamento para minimizar a função de perda (loss). Neste caso, está sendo utilizado o otimizador Adam, que é uma variação do Gradiente Descendente Estocástico (SGD) que incorpora recursos adaptativos de aprendizado para tornar o treinamento mais eficiente.

*   **loss='binary_crossentropy':** A função de perda (ou loss function) é uma medida que quantifica a diferença entre as previsões do modelo e as verdadeiras classes dos dados de treinamento. No caso específico dessa rede neural, está sendo utilizada a função de perda "binary_crossentropy", que é apropriada para problemas de classificação binária, onde a saída deve ser interpretada como probabilidades para as classes.

*   **metrics=['accuracy']:** As métricas são usadas para avaliar o desempenho do modelo durante o treinamento. Neste caso, está sendo utilizado a métrica "accuracy" (acurácia), que é uma medida comum para problemas de classificação. A acurácia mede a proporção de previsões corretas em relação ao total de previsões.

Após a fase de compilação, o modelo está pronto para ser treinado usando os dados de treinamento fornecidos. Durante o treinamento, o otimizador tentará ajustar os pesos da rede para minimizar a função de perda, tornando as previsões do modelo cada vez mais precisas em relação aos dados de treinamento. A acurácia e outras métricas também serão calculadas e exibidas durante o treinamento para monitorar o desempenho do modelo em tempo real.
